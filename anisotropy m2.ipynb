```python
import numpy as np
import matplotlib.pyplot as plt

base_directory = r'C:\Users\firat\OneDrive\Masast\MOKE\anisotropy_measurements_2'
filenames = [base_directory + r'\round2_60degrees', base_directory + r'\round2_90degrees', base_directory + r'\round2_180degrees' ] # Add more filenames as needed


num_curves = len(filenames)  
fig, axes = plt.subplots(num_curves, 1, figsize=(8, 13), sharex=True)


for i, filename in enumerate(filenames):
   
    dataset = np.loadtxt(filename)

    
    axes[i].plot(dataset[:, 0], dataset[:, 1], label='Curve {}'.format(i+1)) #60 DEGREE #90 DEGREE #180 DEGREE
    axes[i].set_ylabel('Y')
    axes[i].set_title('Curve {}'.format(i+1))
    axes[i].grid(True)


axes[-1].set_xlabel('X')


plt.tight_layout()


plt.show()
#60 DEGREE #90 DEGREE #180 DEGREE
```
