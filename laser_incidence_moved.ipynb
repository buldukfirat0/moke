```python
import numpy as np
import matplotlib.pyplot as plt

base_directory = r'C:\Users\firat\OneDrive\Masaüstü\MOKE\kerr_rot_angle\s_polarized'
filenames = [base_directory + r'\ext+5', base_directory + r'\ext-5' ] # Add more filenames as needed


num_curves = len(filenames)  
fig, axes = plt.subplots(num_curves, 1, figsize=(8, 9), sharex=True)


for i, filename in enumerate(filenames):
   
    dataset = np.loadtxt(filename)

    
    axes[i].plot(dataset[:, 0], dataset[:, 1], label='Curve {}'.format(i+1))
    axes[i].set_ylabel('Y')
    axes[i].set_title('Curve {}'.format(i+1))
    axes[i].grid(True)


axes[-1].set_xlabel('X')


plt.tight_layout()


plt.show()
# 25 30 35 40 45
```
