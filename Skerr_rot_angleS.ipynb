```python
import numpy as np
import matplotlib.pyplot as plt

base_directory = r'C:\Users\firat\OneDrive\Masast\MOKE\laser_incidence_moved'
filenames = [base_directory + r'\25', base_directory + r'\30' , base_directory + r'\35' , base_directory + r'\40' , base_directory + r'\45'] # Add more filenames as needed


num_curves = len(filenames)  
fig, axes = plt.subplots(num_curves, 1, figsize=(8, 16), sharex=True)


for i, filename in enumerate(filenames):
   
    dataset = np.loadtxt(filename)

    
    axes[i].plot(dataset[:, 0], dataset[:, 1], label='Curve {}'.format(i+1))
    axes[i].set_ylabel('Y')
    axes[i].set_title('Curve {}'.format(i+1))
    axes[i].grid(True)


axes[-1].set_xlabel('X')


plt.tight_layout()


plt.show()
#ext+5 , ext-5
```
