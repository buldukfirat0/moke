```python
import numpy as np
import matplotlib.pyplot as plt

base_directory = r'C:\Users\firat\OneDrive\Masast\MOKE\anisotropy_measurements'
filenames = [base_directory + r'\15degrees_hysteresis', base_directory + r'\-15degrees_hysteresis', base_directory + r'\30degrees_hysteresis' , base_directory + r'\-30degrees_hysteresis', base_directory + r'\45_degrees_hysteresis' , base_directory + r'\-45degrees_hysteresis', base_directory + r'\60degrees_hysteresis'] # Add more filenames as needed


num_curves = len(filenames)  
fig, axes = plt.subplots(num_curves, 1, figsize=(8, 26), sharex=True)


for i, filename in enumerate(filenames):
   
    dataset = np.loadtxt(filename)

    
    axes[i].plot(dataset[:, 0], dataset[:, 1], label='Curve {}'.format(i+1))
    axes[i].set_ylabel('Y')
    axes[i].set_title('Curve {}'.format(i+1))
    axes[i].grid(True)


axes[-1].set_xlabel('X')


plt.tight_layout()


plt.show()
#-15 +15 DEGREE #30 -30 45 -45 , 60 -60
```
